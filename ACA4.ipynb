### Introduction
The American sitcom *Community* follows a study group at Greendale Community College.  Through the seven group members, the show explores film tropes and complex social scenarios.

*Community* features a very intimate friendship between Troy and Abed, a former high school football star and a socially inept film fanatic.  These friends are beloved by fans, so much so that their dynamic has been called 'Trobed' and discussed as a romantic relationship.  Fans' fascination begs the question: **was Trobed meant to be?**

### Method
In order to determine whether Trobed was an intentional pairing, it is important to analyze their interactions in the first season of the show.  In the first season, characters learn about each other and form bonds that will inform scenarios in later seasons.  In my analysis, I assume that **if the pairing is intentional, then Troy and Abed will more frequently interact with each other in season 1 than with other characters.**

I begin by coding interpersonal interactions in each scene of the first season.  In each scene, a character is said to be involved if they verbally interact with one of the six other main characters.  All the characters involved in each scene are listed and encoded as a series of two-way interactions for that scene.  For example, if one scene includes Britta, Troy, and Abed, the two-way interactions would be Britta / Troy, Britta / Abed, and Troy / Abed.  

All the two-way interactions for the season are added to an adjacency matrix in which the row and column labels are the names of each of the seven characters.  Once the data had been manually entered into the adjacency matrix, a Python script converted the matrix into a frequency table.  This was done so that all edge values would be less than 1.0.

The adjacency matrix was added to Gephi.  The edge width was set to the corresponding value in the adjacency matrix.  The greater the value for a given edge, the greater the edge width.  Such a visualization made it possible to see which characters had a high quantity of interactions with which other characters.

### Results
An examination of Troy's and Abed's node neighborhoods in the social network analysis visualization shows that Troy and Abed were deliberately bonded over the course of the first season of *Community*.  Abed's node network reveals that he most frequently interacted with Jeff and Troy.  Troy's node network reveals frequent interactions with Abed.  Given that the two mutually and consistently bond over the season, significantly more than they do with other characters, **it is likely that the creators of the show deliberately paired the two.  The results suggest that Trobed was meant to be.**

```python
#Import necessary packages
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
#The original dataset contains interpersonal interaction counts
df = pd.read_csv('ACA4_Data.csv', index_col = 0)

columns = ['Troy', 'Shirley', 'Pierce', 'Jeff', 'Britta', 'Annie', 'Abed', 'Total']

#The data is converted to a frequency table
for i in columns:
    df[i] = df[i] / df[i]['Total']
    
df.drop('Total', axis = 1, inplace = True)
df.drop('Total', axis = 0, inplace = True)

#Exported dataset is used to form a SNA visualization in Gephi
df.to_csv('ACA4_FrequencyTable.csv')
print('Frequency Table: Interpersonal Interactions')
df
```
