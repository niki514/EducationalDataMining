```python
#Import necessary packages

import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx.algorithms.community as cm
from networkx.algorithms.community.modularity_max import naive_greedy_modularity_communities
from networkx.algorithms.community.modularity_max import greedy_modularity_communities
```

```python
#Visualizing the network

df = pd.read_csv('ICE8_Data.csv', index_col = 0)
G = nx.Graph(df)
nx.draw(G, with_labels = True)
```

```python
#Network statistics - density
density = nx.density(G)
print("The edge density is %.3f" % density)

#Network statistics - degree centrality
degree = nx.degree(G)
degree = pd.DataFrame(degree, columns = ['Student', '# Edges'])
print("\nNumber of edges per student:")
print(degree)

#Network statistics - closeness centrality
close = nx.closeness_centrality(G)
close = pd.DataFrame.from_dict(close, orient = 'index')
close.reset_index(inplace = True)
close.rename(columns = {'index':'Student', 0:'Closeness Centrality'}, inplace = True)
print("\nCloseness Centrality for each student:")
print(close)

#Network statistics - betweenness centrality
between = nx.betweenness_centrality(G)
between = pd.DataFrame.from_dict(between, orient = 'index')
between.reset_index(inplace = True)
between.rename(columns = {'index':'Student', 0:'Closeness Centrality'}, inplace = True)
print("\nBetweenness Centrality for each student:")
print(between)
```

```python
c = list(naive_greedy_modularity_communities(G))

print(len(c))

community_0 = sorted(c[0])
community_1 = sorted(c[1])
community_2 = sorted(c[2])

print("Community 1:", community_0)
print('\nCommunity 2:', community_1)
print('\nCommunity 3:', community_2)
```
