```python
#Import necessary packages
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Create a scatterplot to look for correlation between HS graduation and college enrollment
df = pd.read_csv("ICE1_Data.csv")
graduation_college = df[['graduation 2010-11', 'college enroll 2010-11']].dropna()
graduation_college.plot.scatter(x = 'graduation 2010-11', y = 'college enroll 2010-11')
```

```python
#Since there is a positive correlation between the two variables, a predictive linear regression model can be made
model = LinearRegression()
graduation = graduation_college['graduation 2010-11'].values.reshape(-1,1)
college = graduation_college['college enroll 2010-11'].values.reshape(-1,1)
model.fit(graduation, college)
print("The intercept is ", model.intercept_)
print("The slope is ", model.coef_)
```

```python
#The linear regression equation is mapped onto the original scatterplot
college_pred = model.predict(graduation)
plt.scatter(graduation, college)
plt.plot(graduation, college_pred, color = 'pink')
plt.xlabel('High School Graduation')
plt.ylabel('College Enrollment')
plt.title('High school graduation is positively correlated with college enrollment')
plt.show()
```

```python
video_data = pd.read_csv("ICE3_Data.csv")
video_data.describe()
video_data.drop(columns = ['stid', 'year'], axis = 1, inplace = True)
pd.plotting.scatter_matrix(video_data, figsize = (15,15))
```

```python
sns.pairplot(video_data)
```

```python
#One way to find a multiple regression model to fit the data
predictors = video_data[['participation', 'confusion.points', 'key.points']].to_numpy()
target = video_data['watch.time'].values.reshape(-1, 1)

video_model = LinearRegression()
video_model.fit(predictors, target)

print("The intercept is ", video_model.intercept_)
print("The coefficients for participation, confusion points, and key points are ", video_model.coef_)
```

```python
#A second way to find a multiple regression model to fit the data
#OLS stands for ordinary least squares - an indicator of model of best fit
from statsmodels.api import OLS
video_model_OLS = OLS(target, predictors)
video_model_OLS = video_model_OLS.fit()
print(video_model_OLS.summary())
```

# Implications of Multiple Linear Regression

As it stands, the linear model to predict watch time is as follows:
watch.time = 0.56 * participation + 2.09 * confusion.points -0.15 * key.points

The results suggest that the time it takes to watch a video is positively correlated with participation and confusion points and negatively correlated with key points.  The more one participates and the greater number of pauses at points of confusion, the more time it takes to watch the video.  The more one fast forwards through the video, the less time it takes to watch the video.

These results do not have any instructional implications, but if the target variable was changed to score on a comprehension assessment, inferences could be made regarding interactions with the video and learning.
