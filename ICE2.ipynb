##### DATA INDEXING IN SERIES

```python
s1 = pd.Series([4, 5, 6, 7], index = ['w', 'x', 'y', 'z'])

print("First element in a series: s1['x'] = ", s1['x'])

print("First two elements in a series: s1['w':'y'] = \n", s1['w':'x'])

print("\nThe loc attribute indexes elements starting with 1, but iloc attribute indexes elements starting with 0.")

s1 = s1.reset_index().drop('index', axis = 1)

print("\ns1.loc[1] = ", s1.loc[1], "\n\ns1.iloc[1] = ", s1.iloc[1])
```

##### DATA INDEXING IN DATAFRAME

```python
df = pd.read_csv("ICE1_Data.csv")

#Categorical variables are not ideal for data analysis.  So, it can be converted to a set of binary variables
#The binary variables consist of the set of values of the original categorical variables

df1 = pd.get_dummies(df['Quality_Review_Score'])
df.drop('Quality_Review_Score', axis = 1, inplace = True)
df = df.join(df1)

#View the first row of the dataframe 
df.loc[0]

#View the first two rows of the dataframe with only DBN, progress report, and student progress scores
df.iloc[:2, :3]
```

##### DATA MASKING

```python
#Locate all schools with a graduation rate above 95%
mask = df['graduation 2010-11'] > 0.95
df[mask]

#Locate all schools with a graduation rate above 95% and a school progress rating below A
mask = (df['graduation 2010-11'] > 0.95) & (df['Progress_Rpt_10-11'] != 'A')
df[mask]
```

##### AGGREGATION, GROUPING, AND PIVOT TABLES

```python
df.mean()

#Total number of schools with each school progress report score
df[['DBN', 'Progress_Rpt_10-11']].groupby('Progress_Rpt_10-11').count()

#Average graduation rate (by school) for each school progress report score
#THIS IS NOT A RATE PER STUDENT
df[['graduation 2010-11', 'Progress_Rpt_10-11']].groupby('Progress_Rpt_10-11').mean()
```

```python
#Making a pivot table showing counts of schools across two dimensions
df.pivot_table('Well Developed', index='Student_Progress_10-11', columns='Progress_Rpt_10-11', aggfunc = np.sum)
```
