```python
#Import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#conda install -c districtdatalabs yellowbrick
from yellowbrick.cluster import SilhouetteVisualizer
from yellowbrick.cluster import silhouette_visualizer
```

```python
#data = two dimensional values for each data point
#cluster = indicates which cluster the data point will belong to
data, cluster = make_blobs(n_samples = 500, centers = 5, cluster_std = 0.5, random_state = 0)
plt.scatter(data[:,0], data[:,1])
```

```python
#k means assigns each data point to one of n clusters
kmeans = KMeans(n_clusters = 5)
kmeans.fit(data)
cluster_kmeans = kmeans.predict(data)
```

```python
#Scatterplot of each datapoint in its cluster, using color as the indicator variable for the cluster
plt.scatter(data[:,0], data[:,1], c = cluster_kmeans, cmap = 'viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c = 'white', s = 100, alpha = 0.5)
```

```python
#Find the silhouette score for each cluster
score = silhouette_score(data, cluster_kmeans, metric = 'euclidean')
print("The silhouette score is ", score)

#Visualize the silhouette score for the five clusters
smodel = KMeans(n_clusters = 5)
visualizer = SilhouetteVisualizer(smodel, colors = 'yellowbrick')
visualizer.fit(data)
visualizer.show()
```

```python
#Alternate method of visualizing silhouette scores
silhouette_visualizer(kmeans, data, colors = 'yellowbrick')
```

```python
#Import the dataset and clean it
#Drop null values
#Drop the column with student identifiers
#Reset row indices to the default
motivation = pd.read_csv('ICE5_Data.csv')
motivation.dropna(inplace = True, axis = 0)
motivation.drop('id', inplace = True, axis = 1)
motivation.reset_index(inplace = True, drop = True)
```

```python
#Try to split the data into two clusters
motivation_kmeans2 = KMeans(n_clusters = 2)
motivation_kmeans2.fit(motivation)
cluster_kmeans2 = motivation_kmeans2.predict(motivation)
score = silhouette_score(motivation, cluster_kmeans2, metric = 'euclidean')
print("The silhouette score is ", score)
silhouette_visualizer(motivation_kmeans2, motivation, colors = 'yellowbrick')
```

```python
#Try to split the data into three clusters
motivation_kmeans3 = KMeans(n_clusters = 3)
motivation_kmeans3.fit(motivation)
cluster_kmeans3 = motivation_kmeans3.predict(motivation)
score = silhouette_score(motivation, cluster_kmeans3, metric = 'euclidean')
print("The silhouette score is ", score)
silhouette_visualizer(motivation_kmeans3, motivation, colors = 'yellowbrick')
```

```python
#Try to split the data into four clusters
motivation_kmeans4 = KMeans(n_clusters = 4)
motivation_kmeans4.fit(motivation)
cluster_kmeans4 = motivation_kmeans4.predict(motivation)
score = silhouette_score(motivation, cluster_kmeans4, metric = 'euclidean')
print("The silhouette score is ", score)
silhouette_visualizer(motivation_kmeans4, motivation, colors = 'yellowbrick')
```
